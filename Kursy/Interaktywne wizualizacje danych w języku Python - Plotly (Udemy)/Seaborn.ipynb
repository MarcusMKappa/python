Wykresy rozkładów - jointplot

```python
sns.set(style='darkgrid')

df = np.random.multivariate_normal(mean=[0, 1], cov=[(1, -0.4), (-0.4, 1)], size=1000)
df = pd.DataFrame(df, columns=['x', 'y'])
df.head(3)
```

```python
df.describe()
```

```python
sns.jointplot(data=df, x='x', y='y')
```

```python
sns.jointplot(data=df, x='x', y='y', kind='hex')
```

```python
#sns.set(style='white')
sns.jointplot(data=df, x='x', y='y', kind='hex')
```

```python
sns.set(style='darkgrid')
sns.jointplot(data=df, x='x', y='y', kind='kde')
```

```python
sns.jointplot(data=df, x='x', y='y', kind='kde', n_levels=15, cmap='viridis')
```

```python
g = sns.jointplot(data=df, x='x', y='y', kind='kde', n_levels=15, cmap='viridis')
g.plot_joint(plt.scatter, c='r', s=30, marker='+')
```
